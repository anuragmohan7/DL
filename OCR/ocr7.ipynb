```python
import cv2
import pytesseract
import re
```

```python
path=pytesseract.pytesseract.tesseract_cmd=r'C:\Users\ASUS\Documents\ocr install\tesseract.exe'
```

```python
img=cv2.imread('AADHAAR.jpg')
cv2.imshow('imagee',img)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
img_grey=cv2.imread('AADHAAR.jpg',cv2.IMREAD_GRAYSCALE)
cv2.imshow('image',img_grey)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
text=pytesseract.image_to_string(img_grey)
print(text)
```

```python
data=pytesseract.image_to_data(img_grey,output_type=pytesseract.Output.DICT)
data
```

```python
n=len(data['text'])
n
```

```python
for i in range(n):
    if re.search(r'(\d{1,2}/\d{1,2}/\d{4})',data['text'][i]):           
        print(data['text'][i])    
```

```python
for i in range(n):
    if re.search(r'(\d{4})',data['text'][i]):           
        print(data['text'][i])    
```

```python
for i in range(n):
        if re.search(r'(\d{4})',data['text'][i]):
                x=data['left'][i]
                y=data['top'][i]
                w=data['width'][i]
                h=data['height'][i]      
                cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,225),2)
cv2.imshow('image',img)
cv2.waitKey(5000)
cv2.destroyAllWindows()         
```
