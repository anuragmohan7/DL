```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('mohanlal2.jpg')
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
#calling inbuild model and creating object
face=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
#face detect using model
faces=face.detectMultiScale(conv_img,minNeighbors=10) #minNeighbors=10 is parameter used to improve performance,10 times machine cheks the face
faces   # here 2 array,  2 face detected
```

```python
#[295,  49, 151, 151],
#  in this first 2 value are x and y cordinate(top left) ,,, 151,151 is width and height,,, inorder to find other coordinate(bottom right)
#           we should add x+w and y+h
for (x,y,w,h) in faces:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(255,0,0),3)
```
