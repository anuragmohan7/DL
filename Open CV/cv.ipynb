```python
#how to convert image to binary
    #color image --> grayscale --> binary

#threshold() is used for binary conversion,  threshold returns 2 outputs
thresh,bin_img=cv2.threshold(img_gray,180,225,cv2.THRESH_BINARY) #thresh is the intencity of black and white (180,225),  bin_img is the image
cv2.imshow('imahgee5',bin_img)
cv2.waitKey(10000)
cv2.destroyAllWindows() 
```

```python
print(thresh)
```

```python
# we can also other function THRESH_OTSU for grayscale, 
thresh,bin_img=cv2.threshold(img_gray,180,225,cv2.THRESH_OTSU) #in THRESH_OTSU black & white is set by machine, user given number is not used
cv2.imshow('imahgee6',bin_img)
cv2.waitKey(10000)
cv2.destroyAllWindows() 
```

```python
#we can use matplotlib for display, because matplotlib has better display than cv2.imshow()
import matplotlib.pyplot as plt
img=cv2.imread('babyimage1.jpg')
plt.imshow(img)
```

```python
#here color is blueish because, while reading image whith cv2 image converted from RGB to BGR,
#  and show image using cv2 image convert from BGR back to RGB,
#  but read image using cv2 but show image using matplotlib becomes blue, because matplotlib doesnot convert backto RGB
```

```python
#to convert BGR to RGB
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
#boxing plot --> used to draw circle,rectangle in an image

#to draw rectangle, 2 points are important ----> left top and right bottom 
cv2.rectangle(conv_img,pt1=(100,18),pt2=(175,100),color=(255,0,0),thickness=2)
plt.imshow(conv_img)
```

```python
#to draw circle, 2 points are important ----> center point and radius
conv_img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

cv2.circle(conv_img2,center=(135,55),radius=40,color=(0,255,0),thickness=2)   #here if you give thickness negative value you can fill color
plt.imshow(conv_img2)
```

```python
# how to add text inside image
conv_img3=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.putText(conv_img3, text='CUTE BABY', org=(180,60), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=0.5, color=(255,0,0), thickness=1)
plt.imshow(conv_img3)
                        # org --> origin of text,  
```
