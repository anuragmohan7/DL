```python
img5=cv2.imread('image3.jpeg')
img5
```

```python
cv2.imshow("image",img5)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
# extracting text
text=pytesseract.image_to_string(img5)
print(text)
```

```python
data=pytesseract.image_to_data(img5,output_type=pytesseract.Output.DICT)
data
```

```python
n=len(data['text'])
n
```

```python
for i in range(n):
    if data['conf'][i]>50: #confidence above 50, means only text
        x=data['left'][i]
        y=data['top'][i]
        w=data['width'][i]
        h=data['height'][i]
        cv2.rectangle(img5,(x,y),(x+w,y+h),(0,0,225),2)
cv2.imshow("image2",img5)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```
