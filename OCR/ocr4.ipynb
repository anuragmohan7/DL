```python

img_grey=cv2.imread('image2.jpeg',cv2.IMREAD_GRAYSCALE)
cv2.imshow('image',img_grey)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
text=pytesseract.image_to_string(img_grey)
print(text)
```

```python
dict=pytesseract.image_to_data(img_grey,output_type=pytesseract.Output.DICT)
dict
```

```python
#DATE
# we use regular expression for extraction of date

import re
```

```python
n=len(dict['text'])
n
```

```python
for i in range(n):
    if re.search(r'(\d{1,2}/\d{1,2}/\d{4})',dict['text'][i]):                #r'(\d{1,2}/\d{1,2}/\d{4})' ==> used to search this pattern
        print(dict['text'][i])                                                                          # r -> read, \d -> digit, \d{1,2} -> may contain 1 or 2 digits
```

```python
# now we can draw rectangle in date
```

```python
for i in range(n):
        if re.search(r'(\d{1,2}/\d{1,2}/\d{4})',dict['text'][i]):
                x=dict['left'][i]
                y=dict['top'][i]
                w=dict['width'][i]
                h=dict['height'][i]      
                cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,225),2)
cv2.imshow('image',img)
cv2.waitKey(5000)
cv2.destroyAllWindows()         
```
